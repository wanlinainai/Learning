```python
from langchain_core.prompts import PromptTemplate

# åˆ›å»ºPromptTemplate
example_prompt = PromptTemplate.from_template(
    template="input:{input}\n, output:{output}"
)

examples = [
    {"input": "åŒ—äº¬å¤©æ°”æ€ä¹ˆæ ·ï¼Ÿ", "output": "åŒ—äº¬å¸‚"},
    {"input": "è±è¥¿å¸‚ä»Šå¤©ä¸‹é›¨å—?", "output": "è±è¥¿å¸‚"},
    {"input": "é’å²›ç°åœ¨çƒ­å—ï¼Ÿ", "output": "é’å²›å¸‚"}
]
few_shot_template = FewShotPromptTemplate(
    example_prompt=example_prompt,
    examples=examples,
    input_variables=["input"],
    suffix="input:{input}\n, output:",
)

few_shot_template.invoke({"input": "å¤©æ´¥ä¼šä¸‹é›¨å—ï¼Ÿ"})

chat_model.invoke(few_shot_template.invoke({"input": "å¤©æ´¥ä¼šä¸‹é›¨å—ï¼Ÿ"}))
```

```python
# 1. å¯¼å…¥ç›¸å…³åŒ…
from langchain_core.prompts import (
    FewShotChatMessagePromptTemplate,
    ChatPromptTemplate
)

# 2. å®šä¹‰ç¤ºä¾‹ç»„
examples = [
    {"input": "2ğŸ¦œ2", "output": "4"},
    {"input": "2ğŸ¦œ3", "output": "8"},
]

# 3. å®šä¹‰ç¤ºä¾‹çš„æ¶ˆæ¯æ ¼å¼æç¤ºè¯æ¨¡ç‰ˆ
example_prompt = ChatPromptTemplate.from_messages([
    ('human', '{input} æ˜¯å¤šå°‘?'),
    ('ai', '{output}')
])

# 4. å®šä¹‰FewShotChatMessagePromptTemplateå¯¹è±¡
few_shot_prompt = FewShotChatMessagePromptTemplate(
    examples=examples,          # ç¤ºä¾‹ç»„
    example_prompt=example_prompt,  # ç¤ºä¾‹æç¤ºè¯è¯æ¨¡ç‰ˆ
)

# 5. è¾“å‡ºå®Œæ•´æç¤ºè¯çš„æ¶ˆæ¯æ¨¡ç‰ˆ
final_prompt = ChatPromptTemplate.from_messages([
    ('system', 'ä½ æ˜¯ä¸€ä¸ªæ•°å­¦å¥‡æ‰'),
    few_shot_prompt,
    ('human', '{input}'),
])

# 6. æä¾›å¤§æ¨¡å‹
import os
import dotenv
from langchain_openai import ChatOpenAI

dotenv.load_dotenv()
os.environ['OPENAI_API_KEY'] = os.getenv("OPENAI_API_KEY")
os.environ['OPENAI_BASE_URL'] = os.getenv("OPENAI_BASE_URL")

chat_model = ChatOpenAI(
    model="gpt-4o-mini",
    temperature=0.4
)

# è°ƒç”¨æ¨¡å‹å¹¶è·å–ç»“æœ
result = chat_model.invoke(final_prompt.invoke(input="2ğŸ¦œ4")).content
print(result)
```
